# Seaborn is a Python library used for data visualization (graphs).
-------------
# It is built on top of Matplotlib and makes graphs more attractive and easier to create.

--------
**Seaborn used for**: bar, histogram, line, scatter, heatmap, boxplot

----------------
**Matplotlib used for**: pie, donut, stacked, area, step, geo

-----------
**Column chart** = Bar chart (same thing)


-----------------
# Why use Seaborn?
Better design (automatic colors & styles)

Works well with datasets (like pandas).

Less code, cleaner output

Built-in themes

# **Bar Chart**

```python
import seaborn as sns
import matplotlib.pyplot as plt

students = ['A','B','C','D']
marks = [50,60,70,80]

sns.barplot(x=students, y=marks)
plt.show()
```

# **Histogram**

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks = [50,60,70,80,60,70,80,90]

sns.histplot(marks, bins=5)
plt.show()
```

# **Column Chart (same as bar)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

students = ['A','B','C','D']
marks = [60,70,80,90]

sns.barplot(x=students, y=marks)
plt.show()
```

# **Line Chart**

```python
import seaborn as sns
import matplotlib.pyplot as plt

students = ['A','B','C','D']
marks = [50,60,70,80]

sns.lineplot(x=students, y=marks)
plt.show()
```

# **Pie chart**

```python
import matplotlib.pyplot as plt

students = ['A','B','C','D']
marks = [50,60,70,80]
plt.pie(
    marks,
    labels=students,
    autopct='%1.1f%%' ,
    shadow=True,
    startangle=90,
    counterclock=True,
)

plt.show()
```

# **Donut Chart**

```python
import matplotlib.pyplot as plt

students = ['A','B','C','D']
marks = [50,60,70,80]

plt.pie(marks, labels=students,autopct='%1.1f%%' ,
    shadow=False,
    startangle=60,
    counterclock=True)
plt.gca().add_artist(plt.Circle((0,0),0.35,color='white'))
plt.show()
```

# **Stacked Bar Chart**

```python
import matplotlib.pyplot as plt

students = ['A','B','C','D']
m = [50,60,70,80]
s = [60,70,80,90]

plt.bar(students, m)
plt.bar(students, s, bottom=m)
plt.show()
```

# **Scatter Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

x = [50,60,70,80]
y = [60,70,80,90]

sns.scatterplot(x=x, y=y)
plt.show()
```

# **Area Chart**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [50,60,70,80]

plt.fill_between(x, y)
plt.plot(x, y)
plt.show()
```

# **Step Chart**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [50,60,70,80]

plt.step(x, y)
plt.show()
```

# **Geo Chart**

```python
import matplotlib.pyplot as plt

x = [72,77,88]   # longitude
y = [23,28,22]   # latitude

plt.scatter(x, y)
plt.show()
```

# **Heatmap**

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [[1,2,3],[4,5,6],[7,8,9]]

sns.heatmap(data, annot=True)
plt.show()
```

# **Box Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [50,60,70,80,90,100]

sns.boxplot(data=data)
plt.show()
```
