```python
import pandas as pd
import numpy as np

cars = pd.Series(np.random.choice(["BMW", "HONDA", "Nissan"], 11))
colors = pd.Series(np.random.choice(["red", "black", "blue", "white"], 11))
odometer = pd.Series(np.random.randint(1000, 200000, 11))
doors = pd.Series(np.random.choice([2, 3, 4, 5], 11))
price = pd.Series(np.random.randint(50000, 2000000, 11))

CAR_SLAES = {
    "Car": cars,
    "Color": colors,
    "Odometer": odometer,
    "Doors": doors,
    "Price": price
}

df = pd.DataFrame(CAR_SLAES )
df["Price"] = "$" + df["Price"].astype(str)
print(df)
df.to_csv("exported_car_sales.csv")
```
